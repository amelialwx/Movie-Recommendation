```python
most_watched_50 = ratings_export.groupby("movie_id").count().sort_values("user_id",  ascending = False).head(50).index
```

```python
most_watched_50 = list(most_watched_50)
```

```python
ratings_export_most_watched = ratings_export[ratings_export["movie_id"].isin(most_watched_50)]
```

```python
ratings_export_most_watched
```

```python
users_vectorized = ratings_export_most_watched.pivot(index= "user_id", columns='movie_id', values='rating_val')
users_vectorized
```

```python
# Trying to fill in NaN values with the average rating of the film 
most_watched_50
```

```python
average_rating_most_watched = ratings_export_most_watched.groupby("movie_id").mean()
average_rating_most_watched["rating_val"] = round(average_rating_most_watched["rating_val"] )
average_rating_most_watched = average_rating_most_watched.reset_index()
average_rating_most_watched
```

```python
dictionary = dict(zip(average_rating_most_watched.movie_id, average_rating_most_watched.rating_val))
dictionary
```

```python
saved = users_vectorized.fillna(value = dictionary)
saved
```

```python
# save 10% of the dataframe for testing 
from sklearn.model_selection import train_test_split

train, test = train_test_split(saved, test_size=0.1)
```

```python
train
```

```python
underrexposed = test.iloc[0]
underrexposed
```

```python
train_data_list = train.values.tolist()
```

```python
from scipy.spatial import distance
import numpy as np
```

```python
# find the top 5 vectors that have the largest consine similary (smallest distance) to the specified vector (user id == "underrexposed")

distances = distance.cdist([underrexposed], train_data_list, "cosine")[0]

res_user = sorted(range(len(distances)), key = lambda sub: distances[sub])[:5]
res_user
```

```python
res_list = list(train.index[res_user])
```
